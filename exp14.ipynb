```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

uploaded = files.upload()

filename = next(iter(uploaded))

img = cv2.imread(filename)
```

```python
rows, cols = img.shape[:2]

src = np.float32([[0,0],[cols-1,0],[0,rows-1],[cols-1,rows-1]])

dst = np.float32([[50,50],[cols-50,20],[20,rows-20],[cols-20,rows-50]])

H, status = cv2.findHomography(src,dst)

homography = cv2.warpPerspective(img,H,(cols,rows))

cv2.imwrite("homography.jpg",homography)

plt.imshow(cv2.cvtColor(homography,cv2.COLOR_BGR2RGB))
plt.title("Homography")
plt.axis("off")
plt.show()
```
