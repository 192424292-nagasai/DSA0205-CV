```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

uploaded = files.upload()

filename = next(iter(uploaded))

img = cv2.imread(filename)
```

```python
rows, cols = img.shape[:2]

src = np.float32([[0,0],[cols-1,0],[0,rows-1],[cols-1,rows-1]])

dst = np.float32([[80,0],[cols-80,0],[0,rows],[cols,rows]])

M = cv2.getPerspectiveTransform(src,dst)

perspective = cv2.warpPerspective(img,M,(cols,rows))

cv2.imwrite("perspective.jpg",perspective)

plt.imshow(cv2.cvtColor(perspective,cv2.COLOR_BGR2RGB))
plt.title("Perspective Transformation")
plt.axis("off")
plt.show()
```
