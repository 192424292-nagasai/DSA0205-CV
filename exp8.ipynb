```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()

filename = next(iter(uploaded))

img = cv2.imread(filename)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Bigger Image
bigger = cv2.resize(img, None, fx=2, fy=2)

# Smaller Image
smaller = cv2.resize(img, None, fx=0.5, fy=0.5)

plt.figure(figsize=(6,6))
plt.imshow(bigger)
plt.title("Bigger Image")
plt.axis("off")
plt.show()

plt.figure(figsize=(4,4))
plt.imshow(smaller)
plt.title("Smaller Image")
plt.axis("off")
plt.show()

cv2.imwrite("bigger.jpg", cv2.cvtColor(bigger, cv2.COLOR_RGB2BGR))
cv2.imwrite("smaller.jpg", cv2.cvtColor(smaller, cv2.COLOR_RGB2BGR))
```
