```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()

filename = next(iter(uploaded))

img = cv2.imread(filename)

clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

counter = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(cv2.cvtColor(clockwise, cv2.COLOR_BGR2RGB))
plt.title("Clockwise Rotation")
plt.axis("off")
plt.show()

plt.imshow(cv2.cvtColor(counter, cv2.COLOR_BGR2RGB))
plt.title("Counter Clockwise Rotation")
plt.axis("off")
plt.show()

cv2.imwrite("clockwise.jpg", clockwise)
cv2.imwrite("counter_clockwise.jpg", counter)
```
